```python
import pandas as pd
import numpy as np

df = pd.read_csv("311_Service_Requests_2020.csv",index_col="Unique Key")

df
```

```python
df["WhetherIllegalParking"] = np.where(df["Complaint Type"] == "Illegal Parking",1,0)
```

```python
global_proportion = df['WhetherIllegalParking'].sum()/df['WhetherIllegalParking'].count()
```

```python
df2 = df.loc[df['Incident Zip'] == 10027]
```

```python
df2["WhetherIllegalParking"] = np.where(df2["Complaint Type"] == "Illegal Parking",1,0)
my_zip_proportion = df2['WhetherIllegalParking'].sum()/df2['WhetherIllegalParking'].count()
```

```python
my_zip_proportion
```

```python
global_proportion
```

```python
higher_parking_proportion = my_zip_proportion > global_proportion
```
